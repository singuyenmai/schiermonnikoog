# Day 1 vs Day 2
## Hare dropping

```r
y11 <- subset(d, Treatment == "Dune" & Day == "1",select = Hare)
y12 <- subset(d, Treatment == "Dune" & Day == "2",select = Hare)

y21 <- subset(d, Treatment == "Salt Marsh" & Day == "1",select = Hare)
y22 <- subset(d, Treatment == "Salt Marsh" & Day == "2",select = Hare)

y31 <- subset(d, Treatment == "Agric. grassland" & Day == "1",select = Hare)
y32 <- subset(d, Treatment == "Agric. grassland" & Day == "2",select = Hare)

A1 <- unlist(y11)
B1 <- unlist(y12)

A2 <- unlist(y21)
B2 <- unlist(y22)

A3 <- unlist(y31)
B3 <- unlist(y32)
```

```r
print("In Dune habitat")
wilcox.test(A1, B1, paired=FALSE) 

print("In Salt Marsh habitat")
wilcox.test(A2, B2, paired=FALSE)

print("In Agric. grassland habitat")
wilcox.test(A3, B3, paired=FALSE)
```

H0: The different days' independent sample distributions are identical

==> Reject H0, samples between days are the same/not significantly different
